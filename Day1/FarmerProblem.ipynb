**Problem Statement**
>A farmer has a goat, a wolf and a cabbage on the west side of a river. He wants to transport all three to the east side using a boat that can carry only the farmer and one additional item. The following constraints must always be satisfied:
- The wolf cannot be left alone with the goat.
- The goat cannot be left alone with the cabbage.

```python
state=[0,0,0,0]
```

```python
# Constraint
def isSafe(state):
    f,g,c,w=state
    if f == 0  :
        if g == 1 and c == 1 :
            return False
        if g == 1 and w == 1:
            return False
    else :
        if g == 0 and c == 0:
            return False
        if g == 0 and w == 0:
            return False
    return True
```

```python
def get_moves(state):
    f,g,c,w = state
    return [
        [1-f, g, c, w],        
        [1-f, 1-g, c, w],     
        [1-f, g, 1-c, w],      
        [1-f, g, c, 1-w]       
    ]
```

```python
def make_moves(state):
    visited = set()
    while state != [1,1,1,1]:
        visited.add(tuple(state))
        for next_state in get_moves(state):
            if isSafe(next_state) and tuple(next_state) not in visited:
                print(next_state)
                state = next_state
                break
        else:
            print("No safe moves left")
            return
```

```python
def solve():
    state = [0,0,0,0]   
    make_moves(state)   
```
