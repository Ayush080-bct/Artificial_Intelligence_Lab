**RobotNavigationProblem**

*A robot is placed in a 5 x 5 grid and must move From starting position(S) to goal state(G) while Avoiding obstacle (x)*

```python
from pprint import pprint
board = [['.']*5 for _ in range(5)]
pprint(board)
```

```python
board[0][0]="S"
board[0][3]="X"
board[1][1]="X"
board[1][3]="X"
board[2][1]="X"
board[3][2]="X"
board[3][3]="X"
pprint(board)
```

```python
def isSafe(board, row, col):
    if row < 0 or row >= len(board) or col < 0 or col >= len(board[0]):
        return False
    if board[row][col] == "X":
        return False
    return True
```

```python
from collections import deque
def bfs(board, start, goal):
    moves = [(-1,0),(1,0),(0,-1),(0,1)]
    queue = deque([[start]])
    visited = set([tuple(start)])

    while queue:
        path = queue.popleft()
        row, col = path[-1]

        if [row, col] == goal:
            print("Shortest path found:")
            for step in path:
                print(step)
            return path

        for dr, dc in moves:
            new_r, new_c = row + dr, col + dc
            if isSafe(board, new_r, new_c) and (new_r, new_c) not in visited:
                visited.add((new_r, new_c))
                queue.append(path + [[new_r, new_c]])
```

```python
goal=[4,4]
start =[0,0]
bfs(board,start,goal)
```
