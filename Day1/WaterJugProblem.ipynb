**Water Jug Problem**
>You are given two water Jugs:
 - Jug A = 4 litres
 - Jug B = 3 liters

Intially both are empty

The objective is to obtain exactly 2 liters of water jug A.

```python
from collections import deque
```

```python
# 1. Generate all possible next states
def get_neighbors(state):
    a, b = state
    neighbors = []

    # Fill A
    neighbors.append((4, b))
    # Fill B
    neighbors.append((a, 3))
    # Empty A
    neighbors.append((0, b))
    # Empty B
    neighbors.append((a, 0))
    # Pour A -> B
    transfer = min(a, 3 - b)
    neighbors.append((a - transfer, b + transfer))
    # Pour B -> A
    transfer = min(b, 4 - a)
    neighbors.append((a + transfer, b - transfer))

    return neighbors
```

```python

# 2. BFS search
def bfs(start, goal_amount):
    queue = deque([[start]])
    visited = set([start])

    while queue:
        path = queue.popleft()
        a, b = path[-1]

        # Goal check
        if a == goal_amount:
            return path

        # Explore neighbors
        for state in get_neighbors((a, b)):
            if state not in visited:
                visited.add(state)
                queue.append(path + [state])

    return None
```

```python
# 3. Solve function
def solve():
    start = (0, 0)
    goal_amount = 2
    solution = bfs(start, goal_amount)

    if solution:
        print("Shortest path to reach 2 liters in Jug A:")
        for step in solution:
            print(step)
    else:
        print("No solution found")
```

```python
solve()
```
