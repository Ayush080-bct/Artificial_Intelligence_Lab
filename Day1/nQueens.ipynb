```python
from  pprint import pprint
n=4
board=[['_']*n for i in range(n)]
pprint(board)   
```

```python
# Placing a queen on row 1 column 1
board[0][0]='Q'
pprint(board)
```

```python
def is_safe(board, row, col, n):
    #column test
    for x in range(row):
       if board[x][col] == 'Q':
           return False
    # top-left Diagonal test
    r = row - 1
    c = col - 1
    while(r>=0 and c>=0):
        if board[r][c] == 'Q':
            return False
        r = r - 1
        c = c - 1
    #Top right Diagonal Test
    r = row - 1
    c = col + 1
    while(r>=0 and c<n):
        if board[r][c] == 'Q':
            return False
        r = r - 1
        c = c + 1
    return True
```

```python
is_safe(board,1,2,n)
```

```python
pprint(is_safe(board, 3, 0, n))
```

```python
def nQueenSolver(board, x, n):
    if x >= n:
        return True
    for col in range(n):
        if is_safe(board, x, col, n):
            board[x][col] = 'Q'
            if nQueenSolver(board, x+1, n):
                return True
            board[x][col]='_'
    return False
```

```python
if nQueenSolver(board, 0, n):
    pprint(board)
```
