```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 2.567
c = 0.123
X = np.arange(10,30,0.13)
y = m*X + c
```

```python
len(X)
```

```python
split = int(len(X)*0.8)
print(split)
X_train, y_train = X[:split], y[:split]
X_test, y_test = X[split:], y[split:]
```

```python
def plot_data(
    X_train=X_train, y_train = y_train, X_test = X_test, y_test = y_test,
    predictions = None):

    plt.scatter(X_train, y_train, c='r', label= 'Training Data')
    plt.scatter(X_test, y_test, c="b", label='Testing Data')
    if predictions is not None:
        plt.scatter(X_test, predictions, c='g', label='Predictions')
    plt.xlabel('X')
    plt.ylabel('y')
    plt.title(' X vs y')
    plt.legend()
    plt.show()
```

```python
plot_data()
```

```python
W = np.random.random()/2
b = np.random.randn()/2
print(f"Intially W = {W} and bias = {b}")
```

```python
y_pred = X_test*W + b
plot_data(predictions = y_pred)
```

```python
def loss_function(y_actual,y_out):
    return np.sum(np.square(y_actual - y_out)/len(y_actual))
```

```python
y_out = X_test*W+ b
plot_data(predictions=y_out)
```

```python
epochs = 200
lr =0.001
train_loss_list = []
test_loss_list = []
for i in range(epochs):
    #forward pass
    train_preds = X_train * W + b
    train_loss = loss_function(y_train, train_preds)
    #compute The gradient
    gradient_W = (-2/len(y_train))*np.sum((y_train - train_preds)*X_train) 
    gradient_b = (-2/len(y_train))*np.sum((y_train - train_preds))
    #Gradient Descent
    W = W - lr * gradient_W
    b = b - lr * gradient_b
    #Evaluate the loss on test set
    test_preds = X_test * W + b
    test_loss =loss_function(y_test, test_preds)
    train_loss_list.append(train_loss)
    test_loss_list.append(test_loss)
    if i%10 == 0:
        print(f"Epoches {i+1} train loss: {train_loss:.4f} and test_loss: {test_loss:.4f}")
```

```python
y_out = X_test*W + b
plot_data(predictions=y_out)
```

```python
plt.plot(range(epochs), train_loss_list, label="Training Loss")
plt.plot(range(epochs), test_loss_list, label="Testing Loss")
plt.title("Epochs vs Loss");
plt.xlabel("Epochs");
plt.ylabel("Loss")
plt.legend();
plt.show()
```
